```python
def convert_examples_to_features(text, label_list, max_seq_length,
                                 tokenizer, label=None):

    label_map = {label: i for i, label in enumerate(label_list)}

    features = []
    

    tokens_a = tokenizer.tokenize(text)


    if len(tokens_a) > max_seq_length - 2:
        tokens_a = tokens_a[:(max_seq_length - 2)]

    tokens = ["[CLS]"] + tokens_a + ["[SEP]"]
    segment_ids = [0] * len(tokens)

    input_ids = tokenizer.convert_tokens_to_ids(tokens)
    input_mask = [1] * len(input_ids)
    seq_length = len(input_ids)

    padding = [0] * (max_seq_length - len(input_ids))
    input_ids += padding
    input_mask += padding
    segment_ids += padding

    assert len(input_ids) == max_seq_length
    assert len(input_mask) == max_seq_length
    assert len(segment_ids) == max_seq_length

    # if output_mode == "classification":
    label_id = label_map[label] if label else None


#     features.append(
#         InputFeatures(input_ids=input_ids,
#                       input_mask=input_mask,
#                       segment_ids=segment_ids,
#                       label_id=label_id,
#                       seq_length=seq_length))
    return input_ids, input_mask, segment_ids, label_id, seq_length
```

```python
def word_concentrated_attention(atts, indexes, tokens, threshold, cross_indexes):
    concentrate_dict = {}
    total_dict = {}
    for i in range(len(atts)):
        atts_elem = atts[i]
        atts_elem = abs(atts_elem)
        
        for i in range(atts_elem.shape[0]):
            for idx in indexes:
                temp = atts_elem[i][idx]
                temp = abs(temp)
                temp = temp/temp.sum()
                #print(temp)
                if temp[idx] > threshold:
                    concentrate_dict[tokens[idx]] = concentrate_dict.get(tokens[idx], 0) + 1
                for cidx in cross_indexes:
                    if temp[cidx] == max(temp):
                        total_dict[tokens[cidx]] = total_dict.get(tokens[cidx], 0) + 1
    return concentrate_dict, total_dict
```

```python
def attention_done(text, threshold = 0.5, is_student=False):
    input_ids, input_mask, segment_ids, label_id, seq_length = convert_examples_to_features(
    text, ['0', '1'], max_seq_length, tokenizer
    )
    
    input_ids = torch.tensor(input_ids)
    input_mask = torch.tensor(input_mask)
    segment_ids = torch.tensor(segment_ids)
    
    input_ids = input_ids[None,:]
    input_mask = input_mask[None,:]
    segment_ids = segment_ids[None,:]
    
    logits, atts, reps = model(input_ids.to(device), segment_ids.to(device), input_mask.to(device), is_student=is_student)
    
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0].detach().cpu().numpy())
    tokens_idx = []
    tokens_cleaned = []
    for i in range(len(tokens)):
        #print(tokens[i])
        if (tokens[i] != '[CLS]') and (tokens[i] != '[SEP]') and (tokens[i] != '[PAD]'):
            tokens_idx.append(i)
            tokens_cleaned.append(tokens[i])
    
    atts_redef = []
    for i in range(len(atts)):
        atts_redef.append(atts[i].detach().cpu().numpy()[0][:,tokens_idx,:][:,:,tokens_idx])
    
    concentrate_dict, total_dict = word_concentrated_attention(atts_redef, range(len(tokens_cleaned)), tokens_cleaned, threshold, range(len(tokens_cleaned)))
    proba = torch.nn.functional.softmax(logits, dim=-1)
    return atts_redef, concentrate_dict, total_dict, tokens_cleaned, proba.detach().cpu().numpy()
```

```python
def plot_per_head(atts_elem, indexes, tokens):
    fig, ax = plt.subplots(atts_elem.shape[0], len(indexes), figsize=(9, 40))
    for i in range(atts_elem.shape[0]):
        for idx in range(len(indexes)):
            temp = atts_elem[i][indexes[idx]]
            temp = abs(temp)
            temp = temp/temp.sum()
            ax[i, idx].bar(tokens, temp)
            ax[i, idx].set_title("self-attention for {}".format(tokens[indexes[idx]]),  size = 'xx-small')
            ax[i, idx].set_ylabel('attention head {}'.format(i + 1))
    
```

```python
def plot_per_attention(atts, indexes, tokens):
    fig, ax = plt.subplots(len(atts), len(indexes), figsize=(9, 40))
    for i in range(len(atts)):
        atts_elem = atts[i]
        atts_elem = abs(atts_elem)
        atts_elem_temp = atts_elem.mean(axis=0)

        for idx in range(len(indexes)):
            temp = atts_elem_temp[indexes[idx]]
            temp = temp/temp.sum()
            ax[i, idx].bar(tokens, temp)
            ax[i, idx].set_title("self-attention for {}".format(tokens[indexes[idx]]),  size = 'xx-small')
            ax[i, idx].set_ylabel('attention layer {}'.format(i + 1))
```
